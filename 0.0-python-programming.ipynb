# Programming with Python
## What is Python?
Python is a programming language---a way of telling a computer what to do.
In particular, Python is an *interpreted, object-oriented, high-level* programming language.
 - an *interpreted* language is a programming language whose programs are translated into machine code by an interpreter at run-time---this is opposed to a *compiled* language, whose programs are translated into machine code *before* run-time.
 - an *object-oriented* programming is a *programming paradigm* based on the concept of `objects`, which can contain data, in the form of fields (attributes), and code, in the form of procedures (methods).
 - actually, programs written in Python are not restricted to the object-oriented paradigm---Python supports *many programming paradigms*, such as object-oriented, imperative, procedural, and functional programming.
 - a *high-level* language is one with strong abstraction from the details of the computer; programs written in high-level languages are usually easier to understand and modify than their low-level counterparts, and hide from the programmer the fine details of how computers actually work.

## The `while` statement: indefinite iteration
The `while` statement is one way of controlling the flow of execution of a program.

```python
while expr:
    suite
else:
    termination suite
other code
```

```python
# sum the integers from 0 to 5
total = 0
value = 0
maximum_value = 5

while value <= maximum_value:
    total = total + value
    value = value + 1
else:
    print(value)

print(total)
```

`while` tests the truth value of `expr`, executing `suite` if `expr` is truthy, and skipping to `other code` if it is falsy.

### Defining functions
Repeating code this way is not only verbose but dangerous.

It is better to write frequently-used code as *functions*.

```python
def function_name(argument_list):  # function definition statement
    suite  # function body
    return return_list  # return statement, optional
```

```python
# function to compute the square of a number
def square(x):
    return x * x
```

```python
square(3)
```

```python
def double(x):
    raise TypeError
    return 2 * x
```

```python
product(3, 4)
```

```python
# function to the norm of two numbers
def norm(x, y, n=2):
    return (x**n + y**n)**(1/n)
```

There are many equivalent ways to call a function

```python
norm(3, 4)
```

```python
norm(3, 4, 2)
```

```python
norm(3, 4, n=2)
```

```python
norm(y=4, n=2, x=3)  # legal, but ugly and confusing
```

```python
norm(n=2, 3, 4)  # SyntaxError
```

```python
norm(x=3, y=4, n=2) # best practice
```

## The `if` statement
`if` statements are another way to control the flow of a program.

```python
if expression1:
    suite1
elif expression2:
    suite2
    ...
elif expressionY:
    suiteY
else:
    suiteZ
other code
```

```python
# fizz buzz
def fizzbuzz(n: int) -> int | str:
    if n % 3 == 0 and n % 5 == 0:
        return "fizzbuzz"
    elif n % 3 == 0:
        return "fizz"
    elif n % 5 == 0:
        return "buzz"
    else:
        return n
```

```python
fizzbuzz(1)
```

```python
fizzbuzz(3)
```

```python
fizzbuzz(5)
```

```python
fizzbuzz(15)
```

### `continue` and `break`
The `continue` and `break` statements are used to implement more complex control flow in loops.

```python
# fizz buzz, but be choosy
n = 0

while n < 20:
    n = n + 1
    fb = fizzbuzz(n)
    if fb == "fizz":
        continue 
    print(n, fb)
    n = n + 1
else:
    print("I didn't break")
```

## Functions of functions
In Python, functions are first-class objects.

They can be passed to and returned from other functions, and names assigned to them, just as with other Python objects.

The type hint for a general function is typing.Callable

`Callable[[*argument_types], *return_type]`

Below, `apply_to_vector` and `reduce_vector` are examples of functions of functions.
This kind of pattern is the backbone of functional programming.

```python
from typing import Callable

def apply_to_vector(f: Callable[[float], float], xs: [float]) -> [float]:
    """
    >>> from math import sqrt
    >>> apply_to_vector(sqrt, [4.0, 9.0, 16.0])
    [2.0, 3.0, 4.0]
    """
    return [f(x) for x in xs]
```

```python
from math import sqrt
xs = [4.0, 9.0, 16.0]
```

```python
apply_to_vector(sqrt, xs)
```

```python
apply_to_vector(lambda x: x**0.5, xs)
```

the builtin function `map` does a similar thing to our `apply_to_vector`.

```python
map(sqrt, xs)
```

```python
list(map(sqrt, xs))
```

```python
def reduce_vector(f: Callable[[float, float], float], xs: [float], init: float) -> float:
    """
    >>> from operator import add
    >>> reduce_vector(add, [4.0, 9.0, 16.0], init=0.0)
    29.0
    """
    if not xs:
        return init
    *xs, x = xs
    return f(x, reduce_vector(f, xs, init))
```

```python
from operator import add
xs = [4.0, 9.0, 16.0]
```

```python
reduce_vector(add, xs, init=0.0)
```

### Example: Lennard-Jones potential
The Lennard-Jones (LJ) potential finds use as a simple model for pair interactions in noble gases and simple fluids, among other applications.

It has the form:
$$
V{\left(r\right)} = 4\epsilon\left[\left(\frac{\sigma}{r}\right)^{12} - \left(\frac{\sigma}{r}\right)^{6}\right] ,
$$
where $\sigma$ is the diameter of the particles and $\epsilon$ is the depth of the potential well.

The total energy of a system of particles interacting *via* the LJ potential is a sum over pairwise contributions,
$$
E = \sum_{i\ne j} V{\left(r_{ij}\right)}
$$
which we can expand to
$$
E\left(\mathbf{x}\right) = \sum_{i\ne j} V{\left( \sqrt{\sum_{\alpha} \left(x_{i\alpha}-x_{j\alpha}\right)^{2}} \right)}
$$
where $i,j$ and $\alpha$ are particle indices a cartesian index, respectively.

With `sum`, `zip`, `combinations`, and comprehensions, we can solve this problem in one line.

```python
from itertools import combinations
from typing import Callable
from math import sqrt


def lj_potential(x: float) -> float:  # obfuscated, but fast implementation!
    r6 = 1 / x / x
    r6 *= r6 * r6
    return 4 * r6 * (r6 - 1)


def pairwise_potential(potential: Callable[[float], float], xss: [[float]]) -> float:
    """
    >>> lj13 = [
    ...     [  1.0132226417,  0.3329955686,  0.1812866397],
    ...     [   0.7255989775, -0.7660449415,  0.2388625373],
    ...     [   0.7293356067, -0.2309436666, -0.7649239428],
    ...     [   0.3513618941,  0.8291166557, -0.5995702064],
    ...     [   0.3453146118, -0.0366957540,  1.0245903005],
    ...     [   0.1140240770,  0.9491685999,  0.5064104273],
    ...     [  -1.0132240213, -0.3329960305, -0.1812867552],
    ...     [  -0.1140234764, -0.9491689127, -0.5064103454],
    ...     [  -0.3513615244, -0.8291170821,  0.5995701458],
    ...     [  -0.3453152548,  0.0366956843, -1.0245902691],
    ...     [  -0.7255983925,  0.7660457628, -0.2388624662],
    ...     [  -0.7293359733,  0.2309438428,  0.7649237858],
    ...     [   0.0000008339,  0.0000002733,  0.0000001488],
    ... ]
    >>> pairwise_potential(lj_potential, lj13)
    -44.326801418734654
    """
    return sum(
        potential(
            sum((xia-xja)**2 for xia, xja in zip(xi, xj))**0.5
        ) for xi, xj in combinations(xss, 2)
    )
```

```python
lj13 = [
    [  1.0132226417,  0.3329955686,  0.1812866397],
    [   0.7255989775, -0.7660449415,  0.2388625373],
    [   0.7293356067, -0.2309436666, -0.7649239428],
    [   0.3513618941,  0.8291166557, -0.5995702064],
    [   0.3453146118, -0.0366957540,  1.0245903005],
    [   0.1140240770,  0.9491685999,  0.5064104273],
    [  -1.0132240213, -0.3329960305, -0.1812867552],
    [  -0.1140234764, -0.9491689127, -0.5064103454],
    [  -0.3513615244, -0.8291170821,  0.5995701458],
    [  -0.3453152548,  0.0366956843, -1.0245902691],
    [  -0.7255983925,  0.7660457628, -0.2388624662],
    [  -0.7293359733,  0.2309438428,  0.7649237858],
    [   0.0000008339,  0.0000002733,  0.0000001488],
]

pairwise_potential(lj_potential, lj13)  # -44.326801418734654
```

What if the function argument *itself* takes arguments?

```python
from typing import Callable


def lj_potential(x: float, epsilon: float, sigma: float) -> float:  # obfuscated, but fast implementation!
    x /= sigma
    r6 = 1 / x / x
    r6 *= r6 * r6
    return 4 * epsilon * r6 * (r6 - 1)


pairwise_potential(lj_potential, lj13)
```

### `*args`, `**kwargs`

```python
from math import dist

def pairwise_potential_with_args(potential: Callable[[float], float], xss: [[float]], **potential_args) -> float:
    return sum(
        potential(dist(xi, xj), **potential_args)
        for xi, xj in combinations(xss, 2)
    )

# potential_args = (1.0, 1.0)
epsilon = 1.0
sigma = 1.0
pairwise_potential_with_args(lj_potential, lj13, epsilon=epsilon, sigma=sigma)
```

### Function factory

```python
def lj_potential_factory(epsilon: float, sigma: float) -> Callable[[float], float]:
    def lj(x: float) -> float:
        return lj_potential(x, epsilon, sigma)
    return lj


lj = lj_potential_factory(epsilon=1.0, sigma=1.0)

pairwise_potential(lj, lj13)
```

### Partial application

```python
from functools import partial

lj = partial(lj_potential, epsilon=1.0, sigma=1.0)

pairwise_potential(lj, lj13)
```

### Custom class
In the below example, the `lj` instance of the `LJPotential` class is not a function object but it *is* callable, because it defines the `__call__` dunder method.

```python
class LJPotential:
    def __init__(self, epsilon: float, sigma: float) -> None:
        self.epsilon = epsilon
        self.sigma = sigma
    
    def __call__(self, x: float) -> float:
        x /= self.sigma
        r6 = 1 / x / x
        r6 *= r6 * r6
        return 4 * self.epsilon * r6 * (r6 - 1)

lj = LJPotential(epsilon=1.0, sigma=1.0)
pairwise_potential(lj, lj13)
```

This approach is a bit more flexible, in that you can later change the values of `epsilon` and `sigma`:

```python
lj.epsilon = 10.0
pairwise_potential(lj, lj13)
```

## Vectorisation
The major benefit of ndarrays is that operations on and between them are vectorised.

Python loops are slow in part because of the global interpreter lock (GIL): even if a loop could be evaluated in parallel, the python interpreter can only look at one element at a time.

Operations like pairwise addition of vectors are parallelisable because the element-wise operations are independent of one another---the result of one operation does not affect another.

Numpy arrays take advantage of this independence, executing multiple operations simultaneously.

Let's apply this to our `pairwise_energy` example.

```python
from itertools import combinations
from typing import Callable
from math import dist


def lj_potential(r: float, epsilon: float=1.0, sigma: float=1.0) -> float:
    r /= sigma
    r6 = 1 / r / r
    r6 *= r6 * r6
    return 4 * epsilon * r6 * (r6 - 1)


def pairwise_energy(xs: [[float]], potential: Callable, **potential_args) -> float:
    return sum(potential(dist(*pair), **potential_args) for pair in combinations(xs, 2))
```

```python
import numpy as np


def lj_potential(r: float | np.ndarray, epsilon: float=1.0, sigma: float=1.0) -> float:
    r /= sigma
    r6 = 1 / r / r
    r6 *= r6 * r6
    return 4 * epsilon * r6 * (r6 - 1)


def pairwise_energy_numpy(xss: [[float]], potential: Callable[[float], float]) -> float:
    xss = np.array(xss)
    atoms, *_ = xss.shape
    left, right = np.triu_indices(atoms, k=1)
    
    return potential(np.linalg.norm(xss[left] - xss[right], axis=1)).sum()
```

## How much faster?

```python
from time import time as time


def time_pairwise(xss, func):
    t0 = time()
    func(xss, lj_potential)
    return time()-t0


atoms = np.geomspace(10,500,25,dtype=int)

random_coordinates = [
    np.random.rand(3*n).reshape(n, 3)
    for n in atoms
]

timings_python = np.array([
    (xss.shape[0],time_pairwise(xss, pairwise_energy))
    for xss in random_coordinates
])

timings_numpy = np.array([
    (xss.shape[0], time_pairwise(xss, pairwise_energy_numpy))
    for xss in random_coordinates
])
```

```python
print(f"{'N':20}{'numpy speedup':20}")
for (n, pytime), (n, npytime) in zip(timings_python, timings_numpy):
    print(f"{n:20n}{pytime/npytime:20.1f}")
```

# Cython: When NumPy is not enough
NumPy excels for vectorisable operations such as element-wise operations.
Reduction operations, such as sums and products, and more complex operations such as computing pairwise energies and forces, involving creating intermediate arrays and can still be slow.

To improve on NumPy's performance in these cases, we can use Cython to compile annotated Python code into `C` functions and classes.

We will use Cython sparingly in this course to optimise small, numerically intensive parts of the code.

When writing Cython code, it is *essential* to explicitly type your variables using type annotations.
Cython is able to bypass the type-checking overhead that comes from Python's dynamic typing by using C types.

int -> cython.int

float -> cython.double

**ensure** that any variables that will be used to index Python objects are annotated as cython.Py_ssize_t---this is the type used in the underlying CPython, and it allows *much* more efficient access to Python object contents.

NumPy arrays should be typed as MemoryViews:

1D np.ndarray(type) -> cython.(type)[::1] ;

2D np.ndarray(type) -> cython.(type)[:, ::1] ; etc

The 1 at the end of '::1' indicates that the array is C-contiguous---you may have to make a copy of your NumPy array before passing it to the function, with the `.copy()` method.

For a NumPy-centric introduction to Cython, see __[here](https://docs.cython.org/en/latest/src/userguide/numpy_tutorial.html)__

To enable Cython compilation in a Jupyter notebook, we use the notebook 'magic' command:

Here an implementation of the pairwise LJ energy in Cython.

```python
%%cython --annotate
# tells Jupyter that this cell should be compiled
# --annotate produces HTML output that shows possible Python interactions in yellow; the fewer, the better!

from cython.cimports.libc.math import sqrt  # allows calling the C function sqrt in the Cython code
import cython  # necessary for type annotations
import numpy as np

@cython.cdivision(True)  # use ONLY if you won't divide by zero
@cython.boundscheck(False)  # use ONLY if you're sure you won't go out of array bounds
@cython.wraparound(False)  # use ONLY if you won't use negative indexing (i.e. vals[-1] to ge tthe last element of vals)
def lj_energy_cython(xss: cython.double[:, ::1], epsilon: cython.double=1.0, sigma: cython.double=1.0)-> float:
    # cython.double[:, ::1] is a MemoryView allowing direct access to the memory underlying an array, bypassing Python overhead
    # ALL variables should be annotated; equivalent to declarations in C and FORTRAN
    # we can still return a Python float
    particles: cython.int = xss.shape[0]  # ALL variables should be annotated; equivalent to declarations in C and FORTRAN
    dimensions: cython.int = xss.shape[1]
    i: cython.Py_ssize_t  # variables used to index arrays should be typed as Py_ssize_t---the type used internally by CPython
    j: cython.Py_ssize_t
    d: cython.Py_ssize_t
    r2: cython.double
    r6: cython.double
    rijd: cython.double
    energy: cython.double = 0.0
    sigma_squared: cython.double = sigma * sigma
    
    # everything else is just normal Python!!
    for i in range(particles):
        for j in range(i + 1, particles):
            r2 = 0.0
            for d in range(dimensions):
                rijd = xss[i, d] - xss[j, d]
                r2 += rijd * rijd
            r6 = sigma_squared / r2
            r6 *= r6 * r6
            energy += r6 * (r6 - 1)
    return 4 * epsilon * energy
```

## How much faster?
To be fair, we should compare with single-function approaches to computing the LJ energy; calling functions incurs overhead

```python
import numpy as np


# this is essentially identical to the Cython version
def lj_energy_python(xss: [[float]], epsilon: float=1.0, sigma: float=1.0) -> float:
    particles = len(xss)
    dimensions = len(xss[0])
    sigma_squared = sigma * sigma
    energy = 0.0
    
    for i in range(particles):
        for j in range(i + 1, particles):
            r2 = 0.0
            for d in range(dimensions):
                rijd = xss[i][d] - xss[j][d]
                r2 += rijd * rijd
            r6 = sigma_squared / r2
            r6 *= r6 * r6
            energy += r6 * (r6 - 1)
    return 4 * epsilon * energy


def lj_energy_numpy(xss: np.ndarray, epsilon: float=1.0, sigma: float=1.0 ) -> float:
    def lj_potential(r) -> float:
        r /= sigma
        r6 = 1 / r / r
        r6 *= r6 * r6
        return 4 * epsilon * r6 * (r6 - 1)
        
    xss = np.array(xss)
    atoms, *_ = xss.shape
    left, right = np.triu_indices(atoms, k=1)    
    return lj_potential(np.linalg.norm(xss[left] - xss[right], axis=1)).sum()
```

Now let's compare times

```python
from time import perf_counter


def time_pairwise(xss, func):
    t0 = perf_counter()
    func(xss)
    return perf_counter()-t0


atoms = np.geomspace(10,1500,25,dtype=int)

random_coordinates = [
    np.random.rand(3*n).reshape(n, 3)
    for n in atoms
]

timings_python = np.array([
    (xss.shape[0],time_pairwise(xss, lj_energy_python))
    for xss in random_coordinates
])

timings_numpy = np.array([
    (xss.shape[0], time_pairwise(xss, lj_energy_numpy))
    for xss in random_coordinates
])

timings_cython = np.array([
    (xss.shape[0], time_pairwise(xss, lj_energy_cython))
    for xss in random_coordinates
])
```

```python
print(f"{'N':20}{'numpy speedup':20}{'cython speedup':20}")
for (n, pytime), (n, npytime), (n, cytime) in zip(timings_python, timings_numpy, timings_cython):
    print(f"{n:20n}{pytime/npytime:20.1f}{pytime/cytime:20.1f}")
```

Use it sparingly!
